```
from pyspark.sql import functions as f
#a.) Write code that to read the csv File?
df = spark.read.format("csv").option("header", "true").option("inferSchema","true").load("/FileStore/tables/Airplane_dataset.csv")  
df.show()
```

```
#b.) How many unique airlines are present in this dataset?
airlines=df.select('UniqueCarrier').distinct().show()
```

```
#c.) Which airline is delayed on departure most often ?
delay_airlines=df.filter(f.col('IsDepDelayed')=='YES').groupBy('UniqueCarrier').count().sort(f.desc('count')).limit(1).show()
```

```
#d.)What was the average arrival delay per airline?
df2=df.withColumn("ArrDelay",f.col("ArrDelay").cast("int"))
avg_delay=df2.filter(f.col('IsArrDelayed')=='YES').groupBy('UniqueCarrier').avg('ArrDelay').select('UniqueCarrier',f.round(f.col("avg(ArrDelay)"),2).alias('avg_delay')).show()
```
